```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

# Initialize WebDriver
driver = webdriver.Chrome()
url = "https://www.techstars.com/newsroom"
driver.get(url)

time.sleep(3)  # Allow the page to load

# Get the entire page source (HTML)
page_html = driver.page_source

# Print the HTML (optional)
print(page_html)

# Save to a file
with open("techstars_newsroom.html", "w", encoding="utf-8") as file:
    file.write(page_html)

print("\n✅ HTML of the first page saved to 'techstars_newsroom.html'")

# Keep the browser open (remove driver.quit() if you want to keep it open)
# driver.quit()
```
