```python
import pandas as pd
from datetime import datetime, timedelta

# Initialize a list to store the article data
article_data = []

# Get the current date and calculate the date 10 days ago
current_date = datetime.now()
ten_days_ago = current_date - timedelta(days=60)

# Define a function to parse the article date
def parse_article_date(date_text):
    try:
        # Parse the article date (adjust the format if necessary)
        return datetime.strptime(date_text, "%Y-%m-%d %H:%M:%S")
    except Exception as e:
        print(f"Error parsing date: {e}")
        return None

# Start from the first article and continue until you reach a date older than 10 days
page_number = 1
i = 1 # Start from article 7 as per your example

while True:
    try:
        # Find the article link
        article_link = driver.find_element(By.XPATH, f'//*[@id="lists"]/li[{i}]/div[1]/h4/a')
        article_url = article_link.get_attribute('href')  # Extract the URL of the article

        # Click the article to go to its page
        article_link.click()
        time.sleep(1)

        # Extract the article's date (adjust the XPath based on the page structure)
        date_text = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div/div[2]/div/ul/li[2]').text

        # Parse the article date
        article_date = parse_article_date(date_text)

        if article_date and article_date >= ten_days_ago:
            # Extract other necessary data (title, view, source, article text)
            title = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div/h1').text
            view = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div/div[2]/div/ul/li[1]').text
            source = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div/div[2]/div/ul/li[3]').text
            article_content = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div/div[2]/article/p[2]').text

            # Append the extracted data to the article_data list
            article_data.append([article_url, title, article_date.strftime('%Y-%m-%d %H:%M:%S'), view, source, article_content])

            # Print extracted data for debugging
            print(f"Article URL: {article_url}")
            print(f"Title: {title}")
            print(f"Date: {article_date.strftime('%Y-%m-%d %H:%M:%S')}")
            print(f"View: {view}")
            print(f"Source: {source}")
            print(f"Article Content: {article_content}")

        # Go back to the list of articles
        driver.back()
        time.sleep(1)

        # Move to the next article
        i += 1

        # If the article is older than 10 days, stop scraping
        if article_date < ten_days_ago:
            print("Reached an article older than 10 days, stopping...")
            break

    except Exception as e:
        print(f"Error extracting data for article {i} on page {page_number}: {e}")
        break  # Stop if there's an error, or you can add a continue statement if you want to skip errors

```

```python


# Initialize a list to store scraped data
article_data = []

# Get the current date and calculate the date 2 months ago
current_date = datetime.now()
two_months_ago = current_date - timedelta(days=20)

# Define function to parse date format correctly
def parse_article_date(date_text):
    try:
        # Adjust the date format if needed
        return datetime.strptime(date_text, "%Y-%m-%d %H:%M:%S")  
    except Exception as e:
        print(f"Error parsing date: {e}")
        return None

# Function to get the current page number (if available)
def get_current_page():
    try:
        active_page = driver.find_element(By.XPATH, '//div[@class="pagination"]/a[@class="active"]')
        return int(active_page.text)
    except:
        return None  # Return None if page number isn't available

# Start scraping from the first page
page_number = 1

while True:  # Loop through pages until we find an old article
    print(f"Scraping page {page_number}...")

    for i in range(1, 51):  # Loop through up to 50 articles per page
        try:
            # Locate the article link
            article_link = driver.find_element(By.XPATH, f'//*[@id="lists"]/li[{i}]/div[1]/h4/a')
            article_url = article_link.get_attribute('href')

            # Click the article to open
            article_link.click()
            time.sleep(1)

            # Extract the article's date
            date_text = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div/div[2]/div/ul/li[2]').text
            article_date = parse_article_date(date_text)

            # Stop scraping if the article is older than 2 months
            if article_date and article_date < two_months_ago:
                print(f"Reached an article older than 2 months ({article_date}), stopping...")
                driver.quit()
                break

            # Extract article details
            title = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div/h1').text
            view = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div/div[2]/div/ul/li[1]').text
            source = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div/div[2]/div/ul/li[3]').text
            article_content = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div/div[2]/article/p[2]').text

            # Append data to list
            article_data.append([article_url, title, article_date.strftime('%Y-%m-%d %H:%M:%S'), view, source, article_content])

            # Print extracted data
            print(f"Scraped: {title} | Date: {article_date}")

            # Go back to the list of articles
            driver.back()
            time.sleep(1)

        except Exception as e:
            print(f"Error extracting data for article {i} on page {page_number}: {e}")
            break  # Stop on error, or use 'continue' to skip errors

    # Check if we should continue scraping or stop
    if article_date and article_date < two_months_ago:
        break

    # Identify the current page number
    current_page = get_current_page()
    
    # Find the "Next Page" button dynamically
    try:
        next_buttons = driver.find_elements(By.XPATH, '//div[@class="pagination"]/a[contains(text(), "Next") or contains(text(), "»")]')

        if next_buttons:
            # Click the last available "Next" button (to avoid duplicates)
            next_buttons[-1].click()
            time.sleep(3)  # Wait for the next page to load
            page_number += 1

            # Verify that we actually moved to a new page
            new_page = get_current_page()
            if new_page is not None and new_page <= current_page:
                print("Warning: The page number didn't increase, stopping to avoid infinite loop.")
                break
        else:
            print("No 'Next Page' button found, stopping pagination.")
            break

    except Exception as e:
        print(f"Error clicking 'Next Page' button: {e}")
        break

# Save data to CSV
```
